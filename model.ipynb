```python
import pandas as pd

data  = pd.read_csv('corporate_mails.csv')
print(data.head(5))

```

```python
urgent_words = [
    "urgent", "asap", "critical", "important",
    "overdue", "escalation", "priority"
]

urgent_phrases = [
    "immediate action required",
    "time sensitive",
    "requires immediate attention",
    "do not delay",
    "respond immediately",
    "urgent request",
    "urgent matter",
    "action needed",
    "attention required"]
```

```python
def clean_text(text):
    text = text.lower()
    text = text.replace(".", "").replace(",", "").replace("/", "")
    return text
```

```python
def compute_urgency_score(email_text):
    # Clean the text
    text = clean_text(email_text)
    
    # Split into words
    words = text.split()
    
    # --- Word-level urgency ---
    urgency_word_count = 0
    for word in words:
        if word in urgent_words:
            urgency_word_count += 1
    
    # --- Phrase-level urgency ---
    urgency_phrase_count = 0
    for phrase in urgent_phrases:
        if phrase in text:
            urgency_phrase_count += 1
    
    # --- Combine ---
    total_urgency_hits = urgency_word_count + urgency_phrase_count
    
    # Normalize by email length
    if len(words) == 0:
        return 0
    
    urgency_score = total_urgency_hits / len(words)
    
    return urgency_score
```

```python
data["urgency_score"] = data["email_body"].apply(compute_urgency_score)
data[["email_body", "urgency_score", "is_bec"]].head(60)
```
